# Model

```python
year = 2019
train = data.loc[data.index.year < year]
test = data.loc[data.index.year == year]

plt.figure(figsize=(15,5))
sns.lineplot(train.index, train['Value'].values, estimator=sum, ci=None);
sns.lineplot(test.index, test['Value'].values, estimator=sum, ci=None);
```

```python
def extract_feat(df, label):
    
    df['date'] = df.index
    df['dayofweek'] = df['date'].dt.dayofweek
    #df['quarter'] = df['date'].dt.quarter,
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['weekofyear'] = df['date'].dt.weekofyear

    X = df[['dayofweek','month','year',
            'dayofyear','weekofyear', 'Border', 'Measure', 'State']]
    if label:
        y = df[label]
        return X, y
    
    return X
```

```python
x_train, y_train = extract_feat(train, 'Value')
x_test, y_test = extract_feat(test, 'Value')

x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
```

```python
xgb = XGBRegressor(n_estimators=1000, n_jobs=-1).fit(x_train, y_train)
np.sqrt(mean_squared_error(y_test, abs(xgb.predict(x_test))))
```

# Results
